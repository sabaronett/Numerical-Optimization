## Booth function

```julia
using Plots
x = y = -10:1:10
f_booth(x, y) = (x + 2*y - 7)^2 + (2*x + y - 5)^2
contourf(x, y, f_booth; fill=:viridis, levels=collect(0:100:2600))
```

## Barnin function

## Rosenbrocks Banana function

```julia
x = -1.5:.1:1.5
y = -2.5:.1:3
a = 1
b = 5
f_banana(x, y) = (a - x)^2 + b*(y - x^2)^2
contourf(x, y, f_banana; fill=:blues, levels=collect(0:2:125))
```

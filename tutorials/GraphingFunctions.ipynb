## The plot function
The most basic usage for plotting a function follows this pattern:

```julia
plot(function_object, from, to) # or plot(f, a, b)
```

```julia
plotly()
plot(sin, 0, 2pi, legend=:outertopright)
```

## Adding layers using "plot!"

```julia
gr()
f(x) = x^2 - 2x + 2
plot(f, -3, 3)
plot!(zero, -3, 3, legend=false)     # x axis
```

## Inf, NaN values (`trim`)

```julia
trim(f; val=10) = x -> abs(f(x)) > val ? NaN : f(x)
```

```julia
f(x) = 1/x
plot(f, -3, 3)
```

```julia
f(x) = 1/x
plot(trim(f), -3, 3)
```

## Plotting with anonymous functions

```julia
function g(x; =/4)
    a = 200cos()
    tan()*x + (32/a)x - 32log(a/(a-x))
end
```

```julia
plot(g, 0, 100)
```

```julia
f(x) = g(x, =/3)
plot(f, 0, 100)
```

```julia
plot(x -> g(x, =/3), 0, 100) # uses anonymous function
```

### For loops

```julia
y = similar(x);
for i in 1:length(x)
    y[i] = f( x[i] )
end
y
```

```julia
x = zeros(Int, 10);        # pre-allocate an int array
x[1:2] = [1,1]
for i in 3:length(x)
    x[i] = x[i-1] + x[i-2]
end
x
```

```julia
i, n = 1, length(x)
while (i <= n)
    print( x[i], " " )
    global i += 1
end
```

## Two functions at once

```julia
plot([sin, cos], 0, 2)
```

```julia
f(x) = x^2
g(x) = 15 + f(x-3)
plot([f, g], -10, 10)
```

```julia
f(x) = x^2 -2
plot([f, x -> 0], -2, 2)
```

```julia
plot(f, -2, 2)
plot!(zero, -2, 2) # zero is a function returning 0, useful in general
```

```julia
function secant(f, a, b)
    m =  (f(b) - f(a)) / (b-a) # slope of secant line
    x -> f(a) + m * (x - a)
end
```

```julia
f(x) = sin(x)
a, b = 0, /2
plot(f, a, b)
plot!(secant(f, a, b), a, b)
```

```julia
f(x) = max(0, 1.0 - abs(x))
function g(f, c, h)
    x -> (1/h)*f((x - c)/h)
end
plot(f, -2, 3)
plot!(g(f, 1, 1/2), -2, 3)
```

```julia
plot([sin, x -> cos(x) > 0 ? 0.0 : NaN], 0, 2)
```
